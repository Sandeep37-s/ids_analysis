```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv(r"C:\Users\sande\Downloads\ids\nsl_kdd\supervised_learning\IDS_Model_Performance.csv")
```

```python
df.head(10)
```

```python
df.set_index('Model', inplace=True)
```

```python
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
```

```python
for metric in metrics:
    plt.figure(figsize=(10, 6))
    df[metric].plot(kind='bar', rot=45, color='skyblue', edgecolor='black')
    plt.title(f'{metric} Comparison Across IDS Models', fontsize=14)
    plt.xlabel('Model')
    plt.ylabel(metric)
    plt.ylim(0, 1)  # All metrics are between 0 and 1
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(12, 6))
```

```python
colors = {
    'Accuracy': 'blue',
    'Precision': 'green',
    'Recall': 'orange',
    'F1-Score': 'red'
}
```

```python
for metric, color in colors.items():
    plt.plot(df.index, df[metric], marker='o', label=metric, color=color, linewidth=2)
```
